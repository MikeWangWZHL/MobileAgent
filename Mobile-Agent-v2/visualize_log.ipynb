```python
import os
import json
import matplotlib.pyplot as plt
import math

def vis_combined_screenshots(screenshot_dir):
    # Gather all screenshot paths and sort them
    screenshot_paths = [os.path.join(screenshot_dir, f) for f in os.listdir(screenshot_dir) if f.endswith('.png') or f.endswith('.jpg')]
    screenshot_paths = sorted(screenshot_paths, key=lambda x: int(os.path.basename(x).split('.')[0]))
    
    # Determine the layout: maximum 6 screenshots per row
    max_columns = 6
    num_screenshots = len(screenshot_paths)
    num_columns = min(num_screenshots, max_columns)
    num_rows = math.ceil(num_screenshots / num_columns)
    
    # Create the figure and axes
    fig, axes = plt.subplots(num_rows, num_columns, figsize=(20, 10))
    
    # Handle single-row or single-column cases
    if num_rows == 1:
        axes = [axes]
    if num_columns == 1:
        axes = [[ax] for ax in axes]
    
    # Plot screenshots
    for i, screenshot_path in enumerate(screenshot_paths):
        row = i // num_columns
        col = i % num_columns
        axes[row][col].imshow(plt.imread(screenshot_path))
        axes[row][col].axis('off')

    # Hide unused axes in the last row
    for j in range(len(screenshot_paths), num_rows * num_columns):
        row = j // num_columns
        col = j % num_columns
        axes[row][col].axis('off')

    plt.tight_layout()
    plt.show()
```
